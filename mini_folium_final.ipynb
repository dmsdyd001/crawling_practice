```python
import folium
```

```python
# 종각역,혜화역,독립문역 위도 경도
latitude = [ 37.569361, 37.5828275 ,37.57441]
longitude = [ 126.9808763, 127.0016805 ,126.95775]

g_map = folium.Map(location=[latitude[0],longitude[0]],zoom_start=18)

g_map
```

```python
import pandas as pd
```

```python
cafe_area_A = pd.read_csv('cafe_list_종각1 (1).csv')
cafe_area_A.head()
```

```python
cafe_area_A.reset_index(inplace=True)
for i in range(len(cafe_area_A)):
    marker01 = folium.Marker([cafe_area_A.loc[i]['위도'], cafe_area_A.loc[i]['경도']],
                       icon = folium.Icon(color='blue', icon='star'))
    marker01.add_to(g_map)
```

```python
g_map
```

```python
franchise = cafe_area_A.loc[cafe_area_A['카페구분'].isin(['프랜차이즈'])]
print(len(franchise))
franchise.head()
```

```python
ind=cafe_area_A.loc[cafe_area_A['카페구분'].isin(['개인카페'])]
ind.head(3)
len(ind)
```

```python
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[latitude[0], longitude[0]],
    zoom_start=15
)

coords_franchise = franchise[['위도', '경도']]
coords_ind = ind[['위도', '경도']]

marker_cluster = MarkerCluster().add_to(m)


for lat, long in zip(coords_ind['위도'], coords_ind['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="red")).add_to(marker_cluster)
for lat, long in zip(coords_franchise['위도'], coords_franchise['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```

```python
#종각역
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[latitude[0], longitude[0]],
    zoom_start=15
)

coords_franchise = franchise[['위도', '경도','카페명','별점']]
coords_ind = ind[['위도', '경도','카페명','별점']]


from folium import Marker
for _, row in coords_franchise.iterrows():
    Marker(location = [row['위도'], row['경도']],
           popup=row['카페명'],tooltip=row['별점'],
           icon=folium.Icon(color='blue',icon='star')
          ).add_to(m)
m
from folium import Marker
for _, row in coords_ind.iterrows():
    Marker(location = [row['위도'], row['경도']],
           popup=row['카페명'],tooltip=row['별점'],
           icon=folium.Icon(color='green',icon='home')
          ).add_to(m)
m
```
